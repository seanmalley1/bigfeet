Combined state agg for ghost, haunting, and bigfeet per capita 

```python
import pandas as pd


bigfeets_df = pd.read_csv('F:\\d10\\Projects\\bigfeet\\data\\export\\bigfeet_capita_simple.csv')
hauntings_df = pd.read_csv('F:\\d10\\Projects\\bigfeet\\data\\export\\ghost_capita_simple.csv')
ufo_df = pd.read_csv('F:\\d10\\Projects\\bigfeet\\data\\export\\ufo_capita_simple.csv')


bigfeets_df.rename(columns={'entries_per_capita': 'bigfeets_per_capita'}, inplace=True)
hauntings_df.rename(columns={'entries_per_capita': 'hauntings_per_capita'}, inplace=True)
ufo_df.rename(columns={'entries_per_capita': 'ufo_per_capita'}, inplace=True)

paranormal_capital = bigfeets_df.merge(hauntings_df, on='state').merge(ufo_df, on='state')

paranormal_capital.rename(columns={
    'count_x': 'bigfeets_count',
    'count_y': 'ghost_count',
    'count': 'ufo_count'
}, inplace=True)

output_path = 'F:\\d10\\Projects\\bigfeet\\data\\export\\combined_capita.csv'
paranormal_capital.to_csv(output_path, index=False)
paranormal_capital.head(50)
```

Bigfoot Calcs.

```python
import pandas as pd
```

```python
df = pd.read_csv('F:\\d10\\Projects\\bigfeet\\data\\bfro_reports_geocoded.csv')
df_pop = pd.read_csv('F:\\d10\\Projects\\bigfeet\\data\\census_state_pop.csv')

state_counts = df['state'].value_counts().reset_index()
state_counts.columns = ['state', 'count']


merged_df = pd.merge(state_counts, df_pop, left_on='state', right_on='NAME')

merged_df['entries_per_capita'] = merged_df['count'] / merged_df['POPESTIMATE2023']
state_counts = merged_df[['state', 'count', 'entries_per_capita']].sort_values(by='entries_per_capita', ascending=False)



df_final = pd.merge(df, state_counts[['state', 'entries_per_capita']], on='state', how='left')


output_path_final = 'F:\\d10\\Projects\\bigfeet\\data\\export\\bigfeet_capita_simple.csv'
state_counts.to_csv(output_path_final, index=False)
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt


df_bf_map = pd.read_csv('F:\\d10\\Projects\\bigfeet\\data\\export\\combined_capita.csv')


geo_df = gpd.read_file('F:\\d10\\Projects\\bigfeet\\data\\states.shp')


bf_merged_df = geo_df.set_index('NAME').join(df_bf_map.set_index('state'))


fig, ax = plt.subplots(1, 1, figsize=(15, 10))
bf_merged_df.plot(column='bigfeets_per_capita', ax=ax, legend=True,
               legend_kwds={'label': "Bigfeets Per Capita by State",
                            'orientation': "horizontal"})

plt.show()

```
